```python
page = requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Head_lines = []
for i in soup.find_all('div', class_= "RiverHeadline-headline RiverHeadline-hasThumbnail"):
    Head_lines.append(i.text)
Head_lines
```

```python
Dates = []
for i in soup.find_all('span', class_= "RiverByline-datePublished"):
    Dates.append(i.text)
Dates
```

```python
news_link = []
for i in soup.find_all('span', class_="LatestNews-container"):
    news_link.append(i.text)
news_link
```
